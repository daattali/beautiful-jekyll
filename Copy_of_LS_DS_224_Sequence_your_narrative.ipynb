## Merge data

https://github.com/pandas-dev/pandas/blob/master/doc/cheatsheet/Pandas_Cheat_Sheet.pdf

```python
# working code to merge income, lifespan, and population
#df = pd.merge(pd.merge(income, lifespan), population)
#df.shape
df = income.merge(lifespan).merge(population)
df.shape
```

this shape telling us  that observation all three numbers

```python
df.head()
```

```python
#we still need reagons and nice country name
entities.head()
```

```python
entities[['country', 'name', 'world_6region']].head()
```

```python
#how do we join this the dataframe we got in progress
df.head()
```

```python
df = pd.merge(df, entities[['country', 'name', 'world_6region']], left_on='geo', right_on='country')
df.shape
```

why there are 8 coulmns now? because add in 3 tables

```python
df.head()
```

```python
df = df.drop(columns=['geo', 'country'])
df.head()
```

just a reminder, our goal is to keep accessible to regular people. 

```python
df = df.rename(columns={'time': 'year','income_per_person_gdppercapita_ppp_inflation_adjusted': 'income', 
                       'life_expectancy_years': 'lifespan', 'population_total': 'population',
                       'name': 'country', 'world_6region': 'region' })
df.tail()
```

## Analyze outliers

```python
now[now['income']>8000].sort_values(by='income')
```

```python
qatar = now[now['country']=='Qatar']
qatar_income = qatar['income'].values[0]
qatar_lifespan = qatar['lifespan'].values[0]
print(qatar_income, qatar_lifespan)
```

```python
sns.relplot(x='income', y='lifespan', hue= 'region', size = 'population', data=now);

plt.text(x=qatar_income -5000, y=qatar_lifespan + 1, s='Qatar')
plt.title('Qatar has the highest income in 2018');
```

```python
years = [1818, 1918, 2018]

centuries = df[df['year'].isin(years)]
centuries.describe()
```

```python
sns.relplot(x='income', y='lifespan', hue= 'region', size = 'population',col='year', data=centuries);

plt.xscale('log');
```

## Point out a story

```python
years = [1918, 1938, 1958, 1978, 1998, 2018]

for year in years:
  sns.relplot(x='income', y='lifespan', hue= 'region', size = 'population', data=df[df['year'] == year])
  
  plt.xscale('log')
  plt.xlim((150, 150000))
  plt.ylim((20,90))
  plt.title(year)
  plt.axhline(y=50, color='grey')
```
